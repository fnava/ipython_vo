```python
# Taken directly from PyVO documentation
object = "M42"

import pyvo as vo
pos=vo.object2pos(object)

print "Searching "+object+" images at position "+str(pos)

survey = "http://skyview.gsfc.nasa.gov/cgi-bin/vo/sia.pl?survey=2massh&"
results = vo.imagesearch(survey, pos=pos, size=0.2, format="image/png")

image = results[0]
print 'Fetching', image.title
image.cachedataset(dir='skyview_images')
```

```python
# Using http://matplotlib.org/users/image_tutorial.html
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img=mpimg.imread('skyview_images/%s.png' % image.title)
imgplot = plt.imshow(img)
```
